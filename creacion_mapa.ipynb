```python
from folium import Map, Marker, Icon, FeatureGroup, LayerControl, Choropleth
from folium.plugins import HeatMap
from folium.vector_layers import Circle, Polygon
import folium
#importamos todas las librerias que nos pueden hacer falta para crear nuestro mapa.
```

```python
m = Map(
    location=[ 47.603122,-122.333253],
    zoom_start=14,
    tiles='Stamen Toner'
)

#creamos el mapa con la localizacion de la oficina con un zoom adecuado, y con el estilo "stamen Toner"
#para que se puedan ver representadas las calles.
```

```python
c = Circle(location=[ 47.603122,-122.333253],radius=1500)
c.add_to(m)
#establecemos mediante un circulo el radio donde se van a encontrar todas las localizaciones alrededor de nuestra nueva 
#oficina.
```

```python
feature_group = folium.FeatureGroup(name="Office")
folium.Marker(location=[ 47.603122,-122.333253]).add_to(feature_group)
m.add_child(feature_group) #establecemos el marcador de la oficina y la añadimos a la leyenda.




folium.Marker(
    location=[47.603122,-122.333253],
    popup='Starbucks',
    icon=folium.Icon(icon='cloud')
).add_to(m) #establecemos el marcador de Starbucks


folium.Marker(
    location=[ 47.607718,-122.334751],
    popup='escuela secundaria Apex Learning Virtual School',
    icon=folium.Icon(color="orange")
).add_to(m)#establecemos el marcador de escuela secundaria y así con todos...

folium.Marker(
    location=[ 47.6011701,-122.3155688],
    popup='escuela primaria Bailey Gatzert Elementary School',
    icon=folium.Icon(color='orange')
).add_to(m)

folium.Marker(
    location=[ 47.6015323,-122.3333255],
    popup='Trinity,club nocturno',
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(m)

folium.Marker(
    location=[ 47.60103050000001,-122.3331875],
    popup='Box House',
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(m)

folium.Marker(
    location=[47.60352899999999,-122.3355838],
    popup='Agencia de viajes Marketplace for safari tours',
    icon=folium.Icon(color='green')
).add_to(m)


folium.Marker(
    location=[47.6061317,-122.3338615],
    popup='Meixi Travel Agency',
    icon=folium.Icon(color='green')
).add_to(m)

folium.Marker(
    location=[ 47.6015323,-122.3333255],
    popup='Jale Chane',
    icon=folium.Icon(color='black')
).add_to(m)

folium.Marker(
    location=[47.60103050000001,-122.3331875],
    popup='Alegre Salon',
    icon=folium.Icon(color='black')
).add_to(m)

#a partir de aqui añadimos a la leyenda todos nuestros marcadores menos el de la propia oficina.
peluqueria1 = FeatureGroup(name='peluqueria Jale Chane')
peluqueria2 = FeatureGroup(name='peluqueria Alegre salon')
starbucks = FeatureGroup(name='Starbucks')
escuela_primaria = FeatureGroup(name="escuela primaria Bailey Gatzert Elementary School")
escuela_secundaria = FeatureGroup(name='escuela secundaria Apex Learning Virtual School')
bar1 = FeatureGroup(name='club Trinity')
bar2 = FeatureGroup(name='club Box House')
a1 = FeatureGroup(name='Agencia de viajes Marketplace for safari tours')
a2 = FeatureGroup(name='Meixi Travel Agency')

folium.Marker(location=[ 47.6015323,-122.3333255]).add_to(peluqueria1)
folium.Marker(location=[47.60103050000001,-122.3331875]).add_to(peluqueria2)
folium.Marker(location=[47.603122,-122.333253]).add_to(starbucks)
folium.Marker(location=[47.6011701,-122.3155688]).add_to(escuela_primaria)
folium.Marker(location=[ 47.607718,-122.334751]).add_to(escuela_secundaria)
folium.Marker(location=[47.6015323,-122.3333255]).add_to(bar1)
folium.Marker(location=[47.60103050000001,-122.3331875]).add_to(bar2)
folium.Marker(location=[47.60352899999999,-122.3355838]).add_to(a1)
folium.Marker(location=[47.6061317,-122.3338615]).add_to(a2)


m.add_child(peluqueria1)
m.add_child(peluqueria2)
m.add_child(starbucks)
m.add_child(escuela_primaria)
m.add_child(escuela_secundaria)
m.add_child(bar1)
m.add_child(bar2)
m.add_child(a1)
m.add_child(a2)

m.add_child(folium.map.LayerControl())
```

```python
#finalmente guardaremos nuestro mapa
m.save("./mapa.html")
```
